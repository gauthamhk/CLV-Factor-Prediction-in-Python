# 2. Introduction

<p style="text-align:justify"> In the modern business world, it is commonly said: “customer is king”. Customer loyalty is one of the influencing factors for firm’s revenue generation. Customer Relationship Management (CRM) is an approach that helps to maintain a healthy business relationship with such customers. CRM helps in improving the customer service relationship of an organization by analyzing customers’ behavioral pattern, building strategies focusing on customer retention and increasing business profits.</p> 

<p style="text-align:justify"> Customer Lifetime Value (CLV) analysis is the prediction of total net income a company can expect from a customer. It is defined as “the net of the revenues obtained from a customer over the lifetime of transactions minus the cost of attracting, selling, and servicing the customer, considering the time value of money”. </p>

<p style="text-align:justify"> Firms focus on the long-term relationship with customers to attain increased profitability. In the context of CRM, CLV or customer equity  is considered as a game changing dimension because it is used as a metric to evaluate the marketing decisions. It can be used for customer selection and marketing resource allocation by developing a framework which is dynamic in nature; managers can proactively maintain/improve customer relationship through marketing contacts across various channels and maximize CLV over time. It is considered as the revenue/money invested by customer minus money invested by the firm to maintain the relationship with the customer. </p>
 
<b> Why do we need to analyze CLV?</b>
 
<p style="text-align:justify"> Understanding and analyzing CLV helps firms in making decisions related to which prospects to acquire, which customers to retain, whom to target for promotion, and how much to spend on acquiring and retaining a customer. Insurance companies see the value of analyzing customer lifetime value because of rapidly commoditizing nature (dominant by third part) of the insurance industry. If sufficient effort is not made by insurers to keep customers at the center of their business, they stand a potential risk of losing a customer to their competitors – who are willing to offer competitive and personalized offers.</p> 

<p style="text-align:justify"> Benefits of CLV analysis in Insurance industry: 
<ul>
  <li>Identifying needs more effectively by understating specific customer requirements</li>
  <li>Increased sales through better timing  by anticipating needs based on historical trends</li>
  <li>Adopt more personalized approach in the development of new products or services</li>
  <li>Design and implement more effective marketing communication for customer needs</li>
  <li>Enhanced customer satisfaction and retention</li>
  <li>Attain increased value from existing customers and reduced cost associated with supporting services; increasing overall          efficiency and reducing total cost of sale</li>
  <li>Identifying profitable customers</li>
  <li>Improved profitability by focusing on most profitable customers and dealing with unprofitable customers in effective ways</li>
</ul></p>
 
<b> How business analytics can help in automization process of CLV analysis/prediction?</b>
 
<p style="text-align:justify"> CLV analysis helps any industry in-understanding customers' preferences, behavior’s, and risks. Implementing business analytics and data science techniques to automate such analysis provides an efficient way to discover hidden behaviors, customer purchase pattern, saves money and time, provides better performance than rules-driven campaigns and pursue a customer-centric economy. The business analytics approach of analyzing CLV, allows marketers to understand what type of subscribers will be the most profitable over a particular time period, reduce the cost of acquisition in the early stages of the relationship, factors aiding the highest return on investment, and implementation of recovery programs.</p>

<p style="text-align:justify">An outlier may indicate bad data. For example, the data may have been coded incorrectly or an experiment may not have been run correctly. If it can be determined that an outlying point is in fact erroneous, then the outlying value should be deleted from the analysis (or corrected if possible). </p>

<p style="text-align:justify">In some cases, it may not be possible to determine if an outlying point is bad data. Outliers may be due to random variation or may indicate something scientifically interesting. In any event, we typically do not want to simply delete the outlying observation. However, if the data contains significant outliers, we may need to consider the use of robust statistical techniques.</p>
